```python
# Import the packages to use

from ipywidgets import interact
import ipywidgets as widgets
```

### Basic Slider

```python
def f(x):
    return x

interact(f, x=10,)
```

### Checkbox

```python
interact(f, x=True)
```

### Textbox

```python
interact(f, x='Hi Nindya!')
```

### Play Button

```python
play = widgets.Play(value=50, min=0, max=100, step=1, description='Tekan Play',
                    disabled=False)
slider = widgets.IntSlider()
widgets.jslink((play, 'value'), (slider, 'value'))
widgets.HBox((play, slider))
```

### Date

```python
widgets.DatePicker(description='Pilih Tanggal', disabled=False)
```

### Tab

```python
tab_contents = ['k0','k1','k2','k3','k4']
children = [widgets.Text(description=name) for name in tab_contents]
tab = widgets.Tab()
tab.children = children
for i in range(len(children)):
    tab.set_title(i, str(i))
tab
```
